## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
# 그래프를 그리는 코드 작성
z=[1,2,3]
w=[2,4,6]

fig , ax = plt.subplots() # 1개의 figure와 1개 이상의 그래프를 생성할 수 있음
ax.plot(x,y)
ax.plot(z,w)

ax.set_title("First plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

## Line Graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label='(x,x)',
    marker='p',
    color='blue',
    linestyle=':'
)
ax.plot(
    x, x**2, label='square',
    marker='2',
    color='red',
    linestyle='--'
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(
    loc='upper left',
    shadow=True,
    fancybox=True,
    borderpad=2
)
```

## scatter

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
     x, x**2, "o",
     markersize=15,
     markerfacecolor='white',
     markeredgecolor="blue"
)

# fig.savefig("plot.png")
# #elice_utils.send_image("plot.png")

x = np.random.randn(50)
y = np.random.randn(50)

#colors = np.random.randint(0,100,50)

sizes = 500 * np.pi * np.random.rand(50) **2
ax.scatter(x,y, s = sizes)
```

## Bar & Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.array(["A", "B", "C", "D", "E"])
y = np.array([18, 7, 12, 10, 8])

#100개의 난수
z = np.random.randn(1000)


fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Bar 그래프
axes[0].bar(x, y)
# 히스토그램
axes[1].hist(z, bins = 50)
```
